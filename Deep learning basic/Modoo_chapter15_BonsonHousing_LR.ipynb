```python
from keras.models import Sequential
from keras.layers.core import Dense 
from sklearn.preprocessing import LabelEncoder 
from sklearn.model_selection import StratifiedKFold

from keras.callbacks import ModelCheckpoint, EarlyStopping

import matplotlib.pyplot as plt
import numpy
import pandas as pd
import tensorflow as tf 
```

```python
seed = 42 
numpy.random.seed(seed)
tf.random.set_seed(seed)
```

```python
df = pd.read_csv("housing2.csv",delim_whitespace=True,header=None)
```

```python
df.head()
```

```python
dataset = df.to_numpy()
```

```python
from sklearn.model_selection import train_test_split
X=dataset[:,0:13].astype(float)
Y=dataset[:,13]
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.3,random_state=seed)
```

```python
model = Sequential()
model.add(Dense(30,input_dim=13,activation='relu'))
model.add(Dense(6,activation='relu'))
model.add(Dense(1)) # Didn't use sigmoid function, because it's regression, not classification
```

```python
model.compile(loss='mean_squared_error',optimizer='adam')
model.fit(X_train,Y_train,epochs=200,batch_size=10)
```

```python
Y_prediction = model.predict(X_test).flatten()
for i in range(10):
    label = Y_test[i]
    prediction = Y_prediction[i]
    print("Actual value : {:.3f}, Predicted value : {:.3f}".format(label,prediction))
```
