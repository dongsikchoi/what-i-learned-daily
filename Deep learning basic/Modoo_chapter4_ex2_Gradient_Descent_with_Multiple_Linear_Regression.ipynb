# Multiple Linear Regression 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
from mpl_toolkits import mplot3d
```

```python
# 일주일간 받는 과외 수업 횟수 변수 추가 
# Add number of times who take private classes in a week 
data = [[2,0,81],[4,4,93],[6,2,91],[8,3,97]]
x1 = [i[0] for i in data] # Study hours 
x2 = [i[1] for i in data] # Number of private class times in a week
y = [i[2] for i in data] # Grade 
```

```python
ax = plt.axes (projection='3d') # Graph type 
ax.set_xlabel('Study hours')
ax.set_ylabel('Private Class')
ax.set_zlabel('Score')
ax.dist = 11
ax.scatter(x1,x2,y)
plt.show()
```

```python
# list to numpy array 
x1_data = np.array (x1)
x2_data = np.array(x2)
y_data = np.array(y)
```

```python
# Initialize gradient value a and intercept value b 

a1 = 0
a2 = 0 
b = 0
```

```python
# learning rate 
lr = 0.05
```

```python
# Number of iterations 
epochs = 2001 
```

```python
# Start gradient descent 
for i in range(epochs):
    y_pred = a1 * x1_data + a2 * x2_data + b # Equation to get y value 
    error = y_data - y_pred # Equation to get error 
    
    a1_diff = -(1/len(x1_data)) * sum(x1_data * (error)) # Partial differentiation with a 
    a2_diff = -(1/len(x2_data)) * sum(x2_data *(error)) # Partial differentiation with a2 
    b_diff = -(1/len(x1_data))*sum(y_data - y_pred ) # Partial differentiation with b 
    
    a1 = a1 - lr * a1_diff # Multiply learning rate to update a1 value 
    a2 = a2 - lr * a2_diff # Multiply learning rate to update a2 value 
    b = b - lr * b_diff # Multiply learning rate to update b value 
    
    if i % 100 == 0:
        print("epoch = %.f, Gradient1=%.04f, Gradient2=%.04f, Intercept=%.04f" 
             % (i,a1,a2,b))
    
    
```
